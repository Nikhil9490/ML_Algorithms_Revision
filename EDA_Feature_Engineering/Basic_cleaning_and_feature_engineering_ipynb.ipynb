**DATA CLEANING**

```python
#missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

open the file and check. there are two datasets(regions) in it with same columns. we can just make a new column based on the region.

```python
dataset.loc[:122, 'region'] = 0
dataset.loc[122:, 'region']= 1
df = dataset
```

```python
df.info()
```

```python
df[['region']]= df[['region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##removing null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fixing space in column names
df.columns= df.columns.str.strip()
df.columns
```

```python
##change required columns as integer data type
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']]= df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## changing the other columns to float datatype
[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
  if i!= 'Classes':
    df[i]= df[i].astype(float)
```
